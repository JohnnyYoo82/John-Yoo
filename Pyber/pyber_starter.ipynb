```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride = pd.merge(ride_data, city_data, on='city', how='left')
```

## Bubble Plot of Ride Sharing Data

```python
#create dataframes for data by city type
urban_cityrides = city_ride[city_ride["type"] == "Urban"].groupby([city_ride["city"]])
suburban_cityrides = city_ride[city_ride["type"] == "Suburban"].groupby([city_ride["city"]])
rural_cityrides = city_ride[city_ride["type"] == "Rural"].groupby([city_ride["city"]])
```

```python
#use to check output
#urban_cityrides.head()
#suburban_cityrides.head()
#rural_cityrides.head()
```

```python
#calculations for Average Fare ($), Total Number of Rides & Total Number of Drivers Per City by City Type (Urban, Suburban, Rural)
urban_avgfare = urban_cityrides["fare"].mean()
urban_totalrides = urban_cityrides["ride_id"].count()
urban_totaldrivers = urban_cityrides["driver_count"].value_counts()

suburban_avgfare = suburban_cityrides["fare"].mean()
suburban_totalrides = suburban_cityrides["ride_id"].count()
suburban_totaldrivers = suburban_cityrides["driver_count"].value_counts()

rural_avgfare = rural_cityrides["fare"].mean()
rural_totalrides = rural_cityrides["ride_id"].count()
rural_totaldrivers = rural_cityrides["driver_count"].value_counts()
```

```python
#use to check output
#urban_avgfare
#urban_totalrides
#urban_totaldrivers
#suburban_avgfare
#suburban_totalrides
#suburban_totaldrivers
#rural_avgfare
#rural_totalrides
#rural_totaldrivers
```

```python
# Obtain the x and y coordinates for each of the three city types & Build the scatter plots for each city types
plt.scatter(urban_totalrides, urban_avgfare, label="Urban", s=20*urban_totaldrivers, color="coral", edgecolor="black", alpha=.5, marker="o")
plt.scatter(suburban_totalrides, suburban_avgfare, label="Suburban", s=20*suburban_totaldrivers, color="lightskyblue", edgecolor="black", alpha=.5, marker="o")
plt.scatter(rural_totalrides, rural_avgfare, label="Rural", s=20*rural_totaldrivers, color="gold", edgecolor="black", alpha=.5, marker="o")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (per city)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
legend = plt.legend(title="City Types", loc="best")
for handle in legend.legendHandles:
    handle.set_sizes([25.0])

# Incorporate a text label regarding circle size
plt.text(45,40, "Note: \nCircle size correlates with driver count per city.")

# Save Figure
plt.savefig("PyberScatterplot.png", bbox_inches = 'tight')

# Show plot
plt.show()
```

## Total Fares by City Type

```python
#create dataframe using merged data by city type
citytype_cityrides = city_ride.groupby(['type'])

# Calculate Type Percents
totalfare = citytype_cityrides['fare'].sum()

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue", "coral"]
explode = (0, 0, 0.1)
plt.pie(totalfare, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=125)

# Incorporate the other graph properties
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("TotalFaresbyCityTypePie.png")

# Show plot
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
totalrides = citytype_cityrides['ride_id'].count()

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue", "coral"]
explode = (0, 0, 0.1)
plt.pie(totalrides, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=125)

# Incorporate the other graph properties
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("TotalRidesbyCityTypePie.png")

# Show plot
plt.show()
```

## Total Drivers by City Type

```python
#create dataframe using city data by city type
citytype_city = city_data.groupby(['type'])

# Calculate Driver Percents
totaldrivers = citytype_city['driver_count'].sum()

# Build Pie Charts
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue", "coral"]
explode = (0, 0, 0.1)
plt.pie(totaldrivers, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=125)

# Incorporate the other graph properties
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("TotalDriversbyCityTypePie.png")

# Show plot
plt.show()
```
